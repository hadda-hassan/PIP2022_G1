# Echantillonage

```python
#On récupère les id et la colonne TOP_FRAUDE
df_FD = data1.loc[:, ('ID', 'TOP_FRAUDE')]
df_FD
```

```python
# df_FD.to_csv("ID_FRAUDE.CSV", index=False)
```

```python
#Somme pas fraude (0)
nb_pas_fraude = (data1['TOP_FRAUDE'] == 0).sum()
nb_pas_fraude
```

```python
#Somme fraude (1)
nb_fraude = (data1['TOP_FRAUDE'] == 1).sum()
nb_fraude
```

```python
#On récupère les lignes pas fraude
pas_fraude = data1.loc[(data1["TOP_FRAUDE"] == 0)]
pas_fraude
```

```python
#On récupère les lignes fraude
fraude = data1.loc[(data1["TOP_FRAUDE"] == 1)]
fraude
```

```python
def echantillon(df_fraude, df_pas_fraude, pourcentage = 0.1):
    ech = df_fraude.sample(frac=pourcentage).append(df_pas_fraude.sample(frac=pourcentage))
    return ech

def train_split_ech(echantillon):
    X = echantillon.drop('TOP_FRAUDE', axis=1)
    y = echantillon['TOP_FRAUDE']
    x_train ,x_test,y_train,y_test=train_test_split(X,y,test_size=0.5)
    return x_train, x_test, y_train, y_test
```

## Echantillon 10%

```python
ech10 = echantillon(fraude, pas_fraude, 0.1)
```

```python
ech10.head()
```

```python
ech10.to_csv("echantillon10.csv", index=False)
```

```python
#Train split
x_train_10,x_test_10,y_train_10,y_test_10=train_split_ech(ech10)
```

```python
x_train_10.to_csv("x_train_10.csv", index=False)
x_test_10.to_csv("x_test_10.csv", index=False)
y_train_10.to_csv("y_train_10.csv", index=False)
y_test_10.to_csv("y_test_10.csv", index=False)
```

## Echantillon 50%

### Visualisation des variables avec un fort taux de données manquantes

```python
colname = df_jeu1_toulouse.columns # Noms des variables du jeu de données

liste_val_abs = []
for col in df_jeu1_toulouse:
    liste_val_abs.append(round(df_jeu1_toulouse[col].isnull().sum() / df_jeu1_toulouse.shape[0]*100)) 
#print(liste_val_abs)

# Récupération des positions des variables avec un nombre de valeurs manquantes supérieur au seuil fixé
#------------------------------------------------------------------------------------------------------
liste_pos = []
seuil = 65
for i in range(0, len(liste_val_abs)):
    if liste_val_abs[i] > seuil:
        liste_pos.append(i) 

print('Nombre de variables avec un taux de valeur manquantes supérieur à :', seuil, '% | effectif =',len(liste_pos))
print()

# Récupération des variables associées
#-------------------------------------
var = []
for i in range(0, len(liste_pos)):
    var.append(colname[liste_pos[i]])
    print(colname[liste_pos[i]],'|',df_jeu1_toulouse[colname[liste_pos[i]]].unique()) 

print()
print('-------------------------------------')
print()

# Nombre de données manquantes pour chaque variable

for i in range(0, len(liste_pos)):
    var.append(colname[liste_pos[i]])
    print(colname[liste_pos[i]], '|',round(df_jeu1_toulouse[colname[liste_pos[i]]].isnull().sum()), '|', round(df_jeu1_toulouse[colname[liste_pos[i]]].isnull().sum() / df_jeu1_toulouse.shape[0]*100),'%')

# On peut séparer les individus qui ont fait des prêts à la banque des individus qui n'en n'ont jamais fait et regarder si ils sont plus fraudés.

# Nombre de variables avec moins de 5% de données manquantes : 246
```
