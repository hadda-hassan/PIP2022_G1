```python
limitPer = len(nom_fichier) * .80
df_sans_na = nom_fichier.dropna(thresh=limitPer, axis=1)
df_sans_na.isna().sum()
```

```python
# Create correlation matrix
corr_matrix = df_sans_na.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] < 0.90)]

# Drop features 
df_sans_na.drop(to_drop, axis=1, inplace=True)
```

```python
corr_df = df_sans_na.corr()


sns.heatmap(corr_df, annot=True)
plt.show()
```

```python
corr_matrix = nom_fichier.corr().abs()

#the matrix is symmetric so we need to extract upper triangle matrix without diagonal (k = 1)

sol = (corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
                  .stack()
                  .sort_values(ascending=False))
sol
#first element of sol series is the pair with the biggest correlation

```
