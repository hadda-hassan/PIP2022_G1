## Creation of dataframes corresponding

```python
df_A = data2[A]
df_A.head()
```

```python
df_B = data2[B]
df_B.head()
```

## Comparison of the variables in the two lists

### Variables common to both lists A and B

```python
C = []
for var in A :
    if var in B :
        C.append(var)
print('Liste des variables communes aux deux listes A et B : ', C)
```

### Correlation of A and B variables

```python
#Calculation of correlations between a variable of A and a variable of B
#Correlation with the standard coefficient: pearson
correl = []
for varA in df_A : 
    for varB in df_B :
        if varA != varB :
            correl.append([varA, varB, df_A[varA].corr(df_B[varB])])
print('CorrÃ©lations entre var de df_A et et var de df_B : ', correl)
```

```python
#Creation of the dataframe of all correlations between A and B
df_corr = pd.DataFrame(correl, columns=['VarA', 'VarB', 'CorrÃ©lation'])
df_corr
```

#### Higher correlations (>0.5) between A and B variables

```python
#Higher correlations (>0.5 in absolute terms)
fort_correl = []
for cor in range(len(correl)):
    if abs(correl[cor][2]) >= 0.5 and abs(correl[cor][2]) <= 1 :
        fort_correl.append(correl[cor])
print('Fortes corrÃ©lations (>0.5) en valeur absolue : ', fort_correl)
```

```python
#Creating a dataframe of higher correlations between a variable of A and a variable of B
df_fort_corr = pd.DataFrame(fort_correl, columns = ['VarA','VarB','CorrÃ©lation'])
df_fort_corr
```

#### Higher correlation for each A variable

```python
#Higher correlation for each A variable
#For each variable of A, we take the variable of B which is the most correlated
max_corr = []
for var in A :
    variable = []
    for i in range(len(correl)) :
        if correl[i][0] == var :
            variable.append(correl[i])
    cor = []
    for i in range(len(variable)) :
        cor.append(variable[i][2])
    for i in range(len(cor)) :
        if cor[i] == max(cor) :
            max_corr.append(variable[i])
print('CorrÃ©lations les plus forte entre une variable de A et celles de B : ', max_corr)
```

```python
#Creating the dataframe
df_max_corr = pd.DataFrame(max_corr, columns=['VarA', 'VarB', 'CorrÃ©lation'])
df_max_corr
```
