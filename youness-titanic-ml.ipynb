# Sex Feautures : 

```python
# Transform Sex Feauture :     1 -> Male    &    0 -> Female
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
sex_encoded = ordinal_encoder.fit_transform(train_data[["Sex"]])
train_data[["Sex"]] = sex_encoded
train_data[["Sex"]] = train_data[["Sex"]].astype(np.uint8)
```

# Embarked Feature : 

# Predict The 2 Missing Values : Random Forest !

```python
Myfeatures = ["Pclass", "Sex", "SibSp", "Parch","Survived","Fare"]
X_embarked = train_data.drop([61,829],axis=0)

X_emabrked_training = pd.get_dummies(X_embarked[Myfeatures])
X_embarked_test = train_data.iloc[[61,829]]
X_embarked_test = pd.get_dummies(X_embarked_test[Myfeatures])
y_embarked = train_data["Embarked"].drop([61,829],axis=0)


from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X_emabrked_training, y_embarked)
predictions = model.predict(X_embarked_test) #Prediction gives Value of 'C'
```

```python
#Replacing Missing Value Of Embarked 
#Later We could Do that Processus in a function 
train_data.loc[61 , "Embarked"] = 'C'
train_data.loc[829, "Embarked"] = 'C'
```

# Embarked : One Hot Encoder
